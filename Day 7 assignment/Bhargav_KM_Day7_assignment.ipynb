##  Inference from the above Analysis:-

##### 1). Most of the Employees are getting less than average salary, (mode < mean).                            
##### 2). Skewness of Education is negative, that means (Mean < Median) .                        
##### 3). Except Education remmaining variables show Positive Skewness, that means (Mean > Median).                
##### 4). Since no variable have 0 Skewness ,all variables are Not normally distributed.         
##### 5). Kurtosis values of Age,Distance, Education, PercentSalaryHike variables are Negative, hence they are Platykurtic(Peakness is low).
##### 6). Kurtosis values of Remmaining variables are Positive, hence they are Leptokurtic(Peakness is high).      
##### 5). STANDARD DEVIATION of MonthlyIncome is HIGH, hence CONSISTENCY  of MonthlyIncome is LOW , hence the company should improve  the Consistency of MonthlyIncome.  
##### 6). Skewness of Age is nearer to 0,hence age is nearly normally distributed.
##### 7). IQR of MonthlyIncome (i.e  83800-29110) is 54690, means there is wide attrition across all income bands.
