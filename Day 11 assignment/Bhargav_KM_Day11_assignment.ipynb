```python
from scipy.stats import pearsonr
```

##  Correlation Calculation for ATTRITION vs Other data :

#### (To show in graph i.e in Scatter plot both the variables must be continuous
####  but here Attrition is not continuous So we cannot implement Using Scatter plot)

### Attrition vs Age:-

```python
stats,p=pearsonr(dataset.Attrition,dataset.Age)
```

```python
print("Stats =",stats,"    and    p =",p)
if p < 0.05:
    print("These two variables are POSITIVELY Correlated and we reject the null hypothesis")
elif p > 0.05:
    print("These two variables are NEGATIVELY Correlated and we accept the null hypothesis")
else :
    print("The variables are Scattered")
```
